# **Parte 1: Fundamentos de bases de datos**

El termino *Conocimiento de Dominio* en la Ciencia de Datos se refiere la experiencia acumulada (conocimiento) que se tiene de una actividad humana o un topico de estudio. El conocimiento de dominio no sólo ayuda a discernir si las respuestas obtenidas mediante un sofisticado análisis estadístico tienen sentido. También es necesario para saber cuáles son las preguntas que deberíamos estar haciendo.

Entre las tareas que un científico de datos puede llevar acabo estan: el desarrollo de estrategias para analizar datos y la preparación de datos para su análisis.

En dichas tareas la limpieza de la base de datos es una actividad clave para prerapar la informacion a tratar. La limpieza de datos también conocida como «Data Cleansing» o «Data Scrubbing» engloba varios procesos destinados a mejorar la calidad de los datos. Estos procesos se utilizan para corregir o eliminar registros inexactos en una base de datos o conjunto de datos. En general, esto significa identificar y sustituir los datos o registros incompletos, inexactos, corruptos o irrelevantes.

Sin la limpieza, es probable que los resultados de los análisis estén distorsionados. Del mismo modo, un modelo de machine learning  o de IA entrenado con datos erróneos puede estar sesgado o ofrecer un rendimiento deficiente.

```python
for index, row in update_df.iterrows():
  for column in update_df[['X6', 'X7', 'X8', 'X9', 'X10', 'X11']]:
    if(not np.isnan(update_df[column][index]) and update_df[column][index] != 0 and update_df[column][index] != -1 and update_df[column][index] < 0):
      update_df.at[index, column] = update_df[column][index] * -1
```

Se convierten los valores diferentes de -1 que son negativos a positivos para las columnas de Historial de pagos pasados

A continuacion se procede a limpiar los valores de las columnas Importe del pago anterior para despues limpiar las columnas de Historial de pagos pasados

```python
for index, row in update_df.iterrows():
  mean = 0
  listColumnsNan = []
  for column in update_df[['X18', 'X19', 'X20', 'X21', 'X22', 'X23']]:
    if(np.isnan(update_df[column][index])):
      listColumnsNan.append(column) #If row has NAN values on previous payment column we get that column
  
  #If has nan we get the mean of this values so we can replace it
  if(len(listColumnsNan) > 0):
    for column in update_df[['X18', 'X19', 'X20', 'X21', 'X22', 'X23']]:
      if(not np.isnan(update_df[column][index])):
        mean = mean + update_df[column][index]

  if(len(listColumnsNan) > 0):
    mean = mean / 6
    for column in listColumnsNan:
      update_df.at[index, column] = mean
```
